```python
import pandas as pd
import zipfile
import os

location = os.path.abspath("C:/Redirected/dsivagnanaretnam/Downloads/endoscopy-video-frames-data.zip")
zf = zipfile.ZipFile(location)
print(zipfile.ZipFile.namelist(zf))

df_1trn = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[1]), header=None, delimiter = ' ')
df_2trn = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[3]), header=None, delimiter = ' ')
df_3trn = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[5]), header=None, delimiter = ' ')
df_4trn = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[8]), header=None, delimiter = ' ')
df_5trn = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[10]), header=None, delimiter = ' ')
df_6trn = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[12]), header=None, delimiter = ' ')
df_7trn = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[6]), header=None, delimiter = ' ')
df_1tst = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[2]), header=None, delimiter = ' ')
df_2tst = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[4]), header=None, delimiter = ' ')
df_3tst = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[0]), header=None, delimiter = ' ')
df_4tst = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[9]), header=None, delimiter = ' ')
df_5tst = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[11]), header=None, delimiter = ' ')
df_6tst = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[13]), header=None, delimiter = ' ')
df_7tst = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[7]), header=None, delimiter = ' ')
```

```python
print(df_1trn.shape)
print(df_2trn.shape)
print(df_3trn.shape)
print(df_4trn.shape)
print(df_5trn.shape)
print(df_6trn.shape)
print(df_7trn.shape)
print(df_1tst.shape)
print(df_2tst.shape)
print(df_3tst.shape)
print(df_4tst.shape)
print(df_5tst.shape)
print(df_6tst.shape)
print(df_7tst.shape)
```

```python
df_1trn.describe().transpose()
```

```python
# Concatenate the training files, replace 255 with 1 and check for missing values
df_train = pd.concat([df_1trn, df_2trn, df_3trn, df_4trn, df_5trn, df_6trn, df_7trn], ignore_index=True)
df_train = df_train.replace(255,1)
df_train.columns = ['Var0', 'Var1', 'Var2', 'Var3', 'Var4', 'Var5', 'Var6', 'Var7', 'Var8', 'Var9', 'Var10', 'Var11', 'Var12', 'Var13', 'Var14', 'Var15', 'Target']
print(df_train.isnull().sum())
df_train
```

```python
df_train['Target'].value_counts()
```

```python
# Create a balanced training dataset
# https://elitedatascience.com/imbalanced-classes

from sklearn.utils import resample

df_train_majority = df_train[df_train.Target==0]
df_train_minority = df_train[df_train.Target==1]
df_minority_upsampled = resample(df_train_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=1871,    # to match majority class
                                 random_state=123) # reproducible results
df_combined = pd.concat([df_train_majority, df_minority_upsampled])
df_combined.Target.value_counts()
```

```python
# Normalise the training dataset
# https://medium.com/datadriveninvestor/k-fold-cross-validation-6b8518070833

from sklearn.preprocessing import MinMaxScaler
from sklearn.utils import shuffle
scaler = MinMaxScaler(feature_range=(0, 1))
df_combined[['Var0', 'Var1', 'Var2', 'Var3', 'Var4', 'Var5', 'Var6', 'Var7', 'Var8', 'Var9', 'Var10', 'Var11', 'Var12', 'Var13', 'Var14', 'Var15']]= scaler.fit_transform(df_combined[['Var0', 'Var1', 'Var2', 'Var3', 'Var4', 'Var5', 'Var6', 'Var7', 'Var8', 'Var9', 'Var10', 'Var11', 'Var12', 'Var13', 'Var14', 'Var15']])
df_combined = shuffle(df_combined)
df_combined
```

```python
df_combined.describe().transpose()
```

```python
X_train_normalised = df_combined.drop(df_combined.columns[16], axis=1)
X_train_normalised
```

```python
Y_train = df_combined[df_combined.columns[16]]
Y_train
```

```python
# Concatenate the testing files, replace 255 with 1 and check for missing values
df_test = pd.concat([df_1tst, df_2tst, df_3tst, df_4tst, df_5tst, df_6tst, df_7tst], ignore_index=True)
df_test = df_test.replace(255,1)
df_test.columns = ['Var0', 'Var1', 'Var2', 'Var3', 'Var4', 'Var5', 'Var6', 'Var7', 'Var8', 'Var9', 'Var10', 'Var11', 'Var12', 'Var13', 'Var14', 'Var15', 'Target']
print(df_test.isnull().sum())
df_test
```

```python
df_test['Target'].value_counts()
```

```python
# Normalise the testing dataset
# https://medium.com/datadriveninvestor/k-fold-cross-validation-6b8518070833

scaler = MinMaxScaler(feature_range=(0, 1))
df_test[['Var0', 'Var1', 'Var2', 'Var3', 'Var4', 'Var5', 'Var6', 'Var7', 'Var8', 'Var9', 'Var10', 'Var11', 'Var12', 'Var13', 'Var14', 'Var15']]= scaler.fit_transform(df_test[['Var0', 'Var1', 'Var2', 'Var3', 'Var4', 'Var5', 'Var6', 'Var7', 'Var8', 'Var9', 'Var10', 'Var11', 'Var12', 'Var13', 'Var14', 'Var15']])
df_test
```

```python
df_test.describe().transpose()
```

```python
X_test_normalised = df_test.drop(df_test.columns[16], axis=1)
X_test_normalised
```

```python
Y_test = df_test[df_test.columns[16]]
Y_test
```

```python
Y_test.value_counts()
```

```python
# https://machinelearningmastery.com/tutorial-first-neural-network-python-keras/


# Create your first MLP in Keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold
import numpy
# fix random seed for reproducibility
numpy.random.seed(7)
N = 10
kf = KFold(n_splits=N)
# create model
model = Sequential()
model.add(Dense(16, input_dim=16, activation='relu'))
model.add(Dense(55, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='mse', optimizer='adam', metrics=['mse','accuracy'])
kfold_number = 0
total_train = 0
total_test = 0
total_train_mse = 0
total_test_mse = 0
number_of_epochs = 800

for train_indices, test_indices in kf.split(X_train_normalised):
    kfold_number += 1
    print("Kfold number is: ", kfold_number)
    model.fit(X_train_normalised.iloc[train_indices], Y_train.iloc[train_indices], epochs=number_of_epochs, batch_size=10)
    scores = model.evaluate(X_train_normalised.iloc[test_indices], Y_train.iloc[test_indices])
    print((model.metrics_names[1], round(scores[1],2)))
    print((model.metrics_names[2], round(scores[2]*100,1)))
    total_train += scores[2]*100
    total_train_mse += scores[1]
    predictions_train = model.predict_classes(X_train_normalised.iloc[test_indices], batch_size=10)
    cm = confusion_matrix(Y_train.iloc[test_indices], predictions_train)
    print(cm)
    print(" ") 
    
average_train_correct_predictions = total_train / N
print("AVERAGE TRAIN: ", round(average_train_correct_predictions,2))
average_train_mse = total_train_mse / N
print("AVERAGE TRAIN MSE: ", round(average_train_mse,2))


model.fit(X_train_normalised, Y_train, epochs=number_of_epochs, batch_size=10)
scores = model.evaluate(X_test_normalised, Y_test)
print((model.metrics_names[1], round(scores[1],2)))
print((model.metrics_names[2], round(scores[2]*100,1)))
#total_test += scores[2]*100
#total_test_mse += scores[1]
predictions_test = model.predict_classes(X_test_normalised, batch_size=10)
cm = confusion_matrix(Y_test, predictions_test)
print(cm)
print(" ")   
```
